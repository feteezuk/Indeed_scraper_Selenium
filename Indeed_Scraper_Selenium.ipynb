## Cell below Specifically tests for "work from home ucsf"

```python
# Get "Work From Home Database"
title = []
company = []
location=[]
link = []

y=10

while y<20: 
    the_link = "https://www.indeed.com/jobs?q=work%20from%20home%20ucsf&start="+str(y)
    the_link = the_link.strip()
    browser.get(the_link)

    #print(the_link)
    
    
    x=0
    while x < len(browser.find_elements_by_tag_name('span[title]')):
        title.append(browser.find_elements_by_tag_name('span[title]')[x].text)
        company.append(browser.find_elements_by_tag_name('a[data-tn-element]')[x].text)
        location.append(browser.find_elements_by_class_name("companyLocation")[x].text)
        link.append(browser.find_elements_by_tag_name('a[data-mobtk]')[x].get_attribute('href'))
        x=x+1
        
    y=y+10    

       

#create DataFrame 
df=pd.DataFrame({'title': title,'company':company, 'location':location, 'link':link})

df
```
